<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(2*-np.pi, 2*np.pi, 200)

tan = np.tan(X)/10
cos = np.cos(X)
sin = np.sin(X)
```

```python
#code here

trig  = [tan,sin,cos]
clr   = ['blue','red','green']
trigs = ['tan','sin','cos']

fig, ax = plt.subplots(1,3,sharey=True,figsize=(12,5))

for i in range(3):
    ax[i].plot(X,trig[i],color=clr[i],label=(trigs[i])+'(x)')
    ax[i].set_title(f'Grafik {trigs[i]}(x)')
    ax[i].set_xticks(np.arange(-6, 7, 2))
        
fig.legend(loc='upper center')
fig.subplots_adjust(top=0.8)
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here

fig, ax = plt.subplots(figsize=(13,8))

ax.plot(x,y)
ax.set_title('Grafik cos(x)')

ins2 = ax.inset_axes([-0.1,-0.1, 0.3, 0.3])
ins3 = ax.inset_axes([ 0.9,-0.1, 0.3, 0.3])
ins4 = ax.inset_axes([ 0.9, 0.9, 0.3, 0.3])
ins5 = ax.inset_axes([-0.1, 0.9, 0.3, 0.3])
ins  = [ins2,ins3,ins4,ins5]
yy   = [y2,y3,y4,y5]
clr2 = ['red','green','black','yellow']

for i in range(4):
    ins[i].plot(x,yy[i],color=clr2[i])
    ins[i].plot(x,y)
    ins[i].set_title(f'Grafik cos(x^{i+2})')
```
