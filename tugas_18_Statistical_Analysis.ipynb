# Soal 2. Persiapan Data

*   Hapus nilai missing dalam rows dataset menggunakan atribut.dropna()

*   Lakukan grouping data berdasarkan kolom 'species' dengan menggunakan agregasi mean()

# Soal 3. Univariate Analysis
Buatlah sebuah analisa univariate dengan menggunakan perintah seperti berikut:

*   Buatlah sebuah Pie plot dari dataset hasil grouping untuk kolom flipper_length_mm seperti expected output berikut:

# Soal 4. Bivariate Analysis

# Soal 5. Multivariate Analisis

*   Dengan menggunakan seaborn, buatlah sebuah pairplot keseluruhan data feature pada dataset (df) dengan argumen kind='bar', seperti expected di bawah:
*   Jelaskan hasil analisa dari multivariate yang dihasilkan



```python
#code here

sns.pairplot(df,kind='scatter') #kind='bar' menghasilkan grafik kosong kecuali di diagonal
plt.show()
```

Expected output:

Pair plot

![](https://drive.google.com/uc?id=1td8WAZLJEW5BPTq-wEJyZ4soRNNLXfqX)

Analisa:

Seperti yang sudah disimpulkan di pertanyaan sebelum ini, berat badan pinguin memiliki korelasi positif terhadap panjang paruh dan panjang *flipper* pinguin. Panjang paruh juga memiliki korelasi positif terhadap panjang *flipper*. Tebal paruh tidak terlihat memiliki korelasi yang jelas terhadap variabel-variabel lain, akan tetapi data tebal paruh dengan panjang *flipper* serta data tebal paruh dengan berat badan yang tervisualisasi seolah menjadi dua set. Untuk analisa lebih lanjut saya membuat *scatter plot* dari dua set variabel tersebut dengan membedakan jenis kelamin dan spesies masing-masing pinguin:

```python
fig, ax = plt.subplots(2,2,sharex=True,figsize=(10,8))

c_sex = {'Male':'Cyan','Female':'Pink'}
c_spc = {'Adelie':'Red', 'Chinstrap':'Green', 'Gentoo':'Blue'}

ax[0][0].scatter('bill_depth_mm','flipper_length_mm',data=df,c=[c_sex[i] for i in df.sex])
ax[0][0].set_ylabel('flipper_length_mm')
ax[0][1].scatter('bill_depth_mm','flipper_length_mm',data=df,c=[c_spc[i] for i in df.species])
ax[1][0].scatter('bill_depth_mm','body_mass_g',data=df,c=[c_sex[i] for i in df.sex])
ax[1][0].set_ylabel('body_mass_g')
ax[1][1].scatter('bill_depth_mm','body_mass_g',data=df,c=[c_spc[i] for i in df.species])
plt.show()
```

Bisa kita lihat bahwa jenis kelamin tidak menambah korelasi antara variabel. Akan tetapi kita bisa lihat bahwa untuk masing masing spesies, terutama untuk spesies *gentoo*, tebal paruh memiliki korelasi positif terhadap panjang *flipper* dan berat badan.
