We see that the headers do not make sense. We drop the first row and drop the first column as it is redundant with the second column.

```python
df = pd.read_csv('Data Market.csv')
df = df.drop(0)
df.reset_index(inplace=True)
df = df.drop(columns=['index','Unnamed: 0'])

df.head()
```

We also move the first 7 headers up by one row by renaming the headers.

```python
new_names = []
old_names = []
for i in range(7):
  new_names.append(df.iloc[1,i])
  old_names.append(df.columns[i])
  df.iloc[1,i] = "NaN"
df.rename(columns=dict(zip(old_names, new_names)), inplace=True)

df.head()
```
