We see that the headers do not make sense. We drop the first row and drop the first column as it is redundant with the second column, and use said second column as the index.

```python
df = pd.read_csv('Data Market.csv')
df = df.drop(0)
df.index = df.loc[:,'Unnamed: 1']
df = df.drop(columns=['Unnamed: 0','Unnamed: 1'])

df.head()
```

We copy the dataframe into two separate variables so we can create a multiindex dataframe

```python
df_noyears = df.iloc[:,:6]

df_noyears
```

```python
df_withyears = df.iloc[:,6:]

print(df_withyears.shape) #if total number of columns !=115 it means that there's an error
df_withyears.head()
```

We then reorganize `df_noyears` by renaming the headers.

```python
new_names = []
old_names = []
for i in range(6):
  new_names.append(df_noyears.iloc[1,i])
  old_names.append(df_noyears.columns[i])
df_headers = df_noyears.rename(columns=dict(zip(old_names, new_names)))

df_headers.index.names = [df_noyears.index[1]]
df_headers = df_headers.drop(index='Ticker').dropna()

df_headers
```
